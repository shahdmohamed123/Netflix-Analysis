```python
actors = dict()
for i in data.cast:
    try:
        stars = i.split(",")
        for star in stars: 
            actors[star] = 0 
    except:
        pass 
```

```python
numberofactors = len(actors)
print(numberofactors)
actors
```

```python
for i in data.cast:
    try:
        stars = i.split(",")
        for star in stars: 
            actors[star] += 1  
    except:
        pass 
```

```python
actors
```

```python
sortedactors =  {k: v for k, v in reversed(sorted(actors.items(), key=lambda item: item[1] ))}
sortedactors
```

```python
top20actors = [ ]
top20number = []
counter= 0 
for actor in sortedactors:
    top20actors.append(actor)
    top20number.append(sortedactors[actor])
    counter+=1 
    if counter == 20 :
        break 

plt.bar(top20actors, top20number)
plt.xticks(rotation= 90)
plt.show()
```
